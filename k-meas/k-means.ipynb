```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Rastgele bir veri seti oluturma
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# DataFrame'e dntrme
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])

# lk birka satr grelim
print(df.head())

# Veriyi grselletirelim
plt.figure(figsize=(8, 6))
plt.scatter(df['Feature1'], df['Feature2'], s=50, alpha=0.7)
plt.title('Rastgele Kme Verisi')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```

```python
# KMeans modelini olutur ve eit
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

# Kme etiketlerini alalm
df['Cluster'] = kmeans.labels_

# Kme merkezlerini yazdralm
print("Kme Merkezleri:")
print(kmeans.cluster_centers_)

# Kmelemeyi grselletirelim
plt.figure(figsize=(8, 6))
plt.scatter(df['Feature1'], df['Feature2'], c=df['Cluster'], cmap='viridis', marker='o', s=100, alpha=0.7)

# Kme merkezlerini de ekleyelim
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Kme Merkezleri')
plt.title('K-Means Kmeleme Sonular')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.legend()
plt.show()
```
