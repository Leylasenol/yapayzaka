```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Rastgele bir veri seti oluşturma
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# DataFrame'e dönüştürme
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])

# İlk birkaç satırı görelim
print(df.head())

# Veriyi görselleştirelim
plt.figure(figsize=(8, 6))
plt.scatter(df['Feature1'], df['Feature2'], s=50, alpha=0.7)
plt.title('Rastgele Küme Verisi')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```

```python
# KMeans modelini oluştur ve eğit
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

# Küme etiketlerini alalım
df['Cluster'] = kmeans.labels_

# Küme merkezlerini yazdıralım
print("Küme Merkezleri:")
print(kmeans.cluster_centers_)

# Kümelemeyi görselleştirelim
plt.figure(figsize=(8, 6))
plt.scatter(df['Feature1'], df['Feature2'], c=df['Cluster'], cmap='viridis', marker='o', s=100, alpha=0.7)

# Küme merkezlerini de ekleyelim
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Küme Merkezleri')
plt.title('K-Means Kümeleme Sonuçları')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.legend()
plt.show()
```
