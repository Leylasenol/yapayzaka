```python
import pandas as pd
#veri setini oluşturma
data={
    'Year':[2020,2021,2022,2023,2024],
    'Sales (USD)':[50000,75000,100000,120000,150000],
    'Customers':[120,180,250,300,400],
}
df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(12,6))
```

```python
#Satışların çizgi grafiği
plt.subplot(1,2,1)
plt.plot(df['Year'],(df['Sales (USD)'],marker='o',color='black')
plt.title('Yıllık Satışlar')
plt.xlabel('Yıl')
plt.ylabel('Sales (USD)')
```
